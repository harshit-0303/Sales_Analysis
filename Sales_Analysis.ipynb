```python
import pandas as pd
import numpy as np
df = pd.read_csv(r"D:\New folder (2)\python_dsa\ecommerce_data.csv")
```

```python
df.head()
```

# Section 1 : Basic Pandas

```python
# DataTypes of each column 
df.info()
```

```python
# Shape of data
df.shape
```

```python
# Convert 'OrderDate' to datetime

df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['OrderID']= df['OrderID'].astype(object)
```

# Section 2: Data Cleaning 


```python
# Missing Values
df.isnull().sum()
```

```python
# Dropping Duplicates

df.drop_duplicates()
```

```python
# Replacing values

df.replace(np.where(df['Rating']<3),'Nan',inplace=True)
```

```python
df['Rating'].dtype
```

```python
# Convert object datatype to numeric
df['Rating'] = pd.to_numeric(df['Rating'],errors = 'coerce')
```

```python
# Filling missing values

df['Rating']=df['Rating'].fillna(df['Rating'].mean())
```

# Section 3 : Filtering

```python
df[df['City'] == "Bangalore"]
```

```python
df[df['Age']>30]
```

# Section 4 : Feature Engineering

```python
# Creating new column
df['TotalAmount'] = df['Price']* df['Quantity']
```

```python
# Creating column based on condition 
conditions = [
    ((df['Age']>20) & (df['Age'] < 30)),
    ((df['Age'] >= 30) & (df['Age'] < 40)),
    (df['Age']>= 40 )
]
choices = ['Young','Adult','Senior']
df['Agegroup'] = np.select(conditions,choices,default = "Missing")
```

# Section 5: GroupBy

```python
revenue_by_city = df.groupby('City')['TotalAmount'].sum()
revenue_by_city
```

```python
avg_rating_per_category = df.groupby('Category')['Rating'].mean().reset_index(name = 'Avg_rating')
avg_rating_per_category
```

```python
orders_per_payment_method = df.groupby('PaymentMethod')['OrderID'].count().reset_index(name= 'count')
orders_per_payment_method
```

# Section 6 : Advanced Analysis

```python
top_3_product = (
    df.groupby('Product')['TotalAmount']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
    .head(3)
)
top_3_product
```

```python
most_popular_category = (
    df.groupby('Category')['Quantity']
    .sum()
)
most_popular_category
```

```python
city_highest_revenue = (
    df.groupby('City')['TotalAmount']
    .sum()
    .sort_values(ascending = False)
    .reset_index()
    .head(1)
)
city_highest_revenue
```

# Section 7 : Sorting & Ranking

```python
top_5_customers = (
    df.groupby('CustomerID',as_index=False)
      .agg({
          'TotalAmount': 'sum',
          'Age': 'first',
          'City': 'first'
      })
      .sort_values(by='TotalAmount', ascending=False)
      .head(5)
)

top_5_customers
```

```python
df['Rank'] = df['TotalAmount'].rank().astype(int)
```

```python
# Ranking cities based on Sales

rank_cities = (
    df.groupby("City")['TotalAmount']
    .sum()
    .rank(method='dense',ascending=False)
    .astype(int)
    .reset_index(name = "Rank")
    .sort_values(by='Rank')

)

rank_cities
```

## SECTION 8: Window Functions

```python
df.columns
```

```python
# Running total of revenue

df['running_total'] = df.sort_values('OrderDate')['TotalAmount'].cumsum()
df['running_total']
```

```python
# Rolling average of price (window=3)

df['rolling_avg'] = df.sort_values("OrderDate")['Price'].rolling(window=3).mean()
df[['OrderDate', 'rolling_avg']]
```

```python
# Rank orders within each city
df['CityRank'] = (
    df.groupby('City')['TotalAmount']
    .rank(method='dense',ascending=False)
    .astype(int)
)
df[['City', 'TotalAmount', 'CityRank']].sort_values(by=['City', 'CityRank'])    
```

# SECTION 9: NumPy

```python
# Convert Price column to NumPy array
price = np.array(df['Price'])
price
```

```python
# Find: Mean,Median,Standard deviation
price_mean = round(np.mean(price),2)
price_median = round(np.median(price),2)
price_std = round(np.std(price),2)
print(f"Mean: {price_mean}, Median: {price_median}, Standard Deviation: {price_std}")
```

```python
# Normalize price values
df['Price_normalized'] = (df['Price'] - price_mean) / price_std
df['Price_normalized']
```

```python
# Find outliers using Z-score
z_scores = df['Price_normalized']
outliers = df[np.abs(z_scores) > 3]
outliers
```


# SECTION 10: Real Business Questions

```python
# Which category generates highest revenue?

category_revenue = df.groupby('Category')['TotalAmount'].sum().sort_values(ascending=False)
category_revenue
```

```python
# Which payment method is most used?

payment_method_counts = df['PaymentMethod'].value_counts()
payment_method_counts
```

```python
# Do higher prices lead to better ratings?

price_rating_correlation = df['Price'].corr(df['Rating'])
price_rating_correlation
```

```python
# Which age group spends the most?
spent_by_age_group = df.groupby('Agegroup')['TotalAmount'].sum()
spent_by_age_group
```

```python
# Which city has lowest average rating?

lowest_avg_rating = df.groupby('City')['Rating'].mean().sort_values().reset_index(name='Avg_Rating')
lowest_avg_rating
```
